Blackbody energy distribution curve for a given temperature

```python
import numpy as np
import matplotlib.pyplot as plt
from astropy import constants as const
from astropy import units as u
```

```python
wavelengths=np.arange(100,30000,100)*u.angstrom
T=5000*u.K
def calc_radiated_power(wavelengths,T):
    coeff=(2*np.pi*(const.c**2)*const.h)
    var=(1/wavelengths**5)*(1/((np.exp((const.c*const.h)/(wavelengths*const.k_B*T)))-1))
    return coeff*var
```

```python
radiated_power_5000=calc_radiated_power(wavelengths,T)
T=12500*u.K
radiated_power_12500=calc_radiated_power(wavelengths,T)
```

```python
plt.plot(wavelengths,radiated_power_5000)
```

```python
plt.plot(wavelengths,radiated_power_12500)
```
